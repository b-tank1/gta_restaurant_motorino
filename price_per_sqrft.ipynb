```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
neighborhood_locations = pd.read_csv(Path('C:/Users/chefd/OneDrive/Desktop/locations_price_per_square_foot.csv')).set_index('Neighbourhood')
```

```python
neighborhood_locations
```

```python
neighborhood_locations.drop(columns=['Address'], inplace=True)
```

```python
neighborhood_locations
```

```python
neighborhood_locations.hvplot.points('Lon', 
                                   'Lat', 
                                   geo= True, 
                                   color= "red",  
                                   size = 'Price per square foot', 
                                   #color='Price per square foot', 
                                   frame_width = 700, 
                                   frame_height = 500,
                                   tiles= "ESRI",  
                                   title= 'Potential Locations For New Restataurant'
                                  )
```

```python
neighborhood_locations.hvplot.points('Lon', 
                                   'Lat', 
                                   geo= True, 
                                   color= 'red',  
                                   alpha= 0.9,  
                                   size = 'Price per square foot',
                                   #color='Price per square foot', 
                                   ylabel= "Lat",
                                   xlabel= "Lon",  
                                   frame_width = 900, 
                                   frame_height = 700,
                                   tiles= "OSM",  
                                   title= 'Potential Locations For New Restataurant'
                                  )
```
